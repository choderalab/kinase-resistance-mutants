```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from matplotlib.offsetbox import AnchoredText
import matplotlib.cm as cm
import pandas as pd
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import FormatStrFormatter
import matplotlib
import assaytools
from assaytools import platereader
```

```python
filenames = [
    'Abl_I403T_Bos_20180808_112442.xml', 
    'Abl_I403T_Erl_20180808_120212.xml', 
    'Abl_I403T_Gef_20180808_125307.xml', 
    'Abl_I403T_Pon_2010808_131919.xml'
]

datapath = 'data/I403T'
label = 'ex280_scan_top_gain100'
```

```python
reads = list()
nreads = 4
for read in range(nreads):
    read = platereader.read_icontrol_xml(os.path.join(datapath, filenames[read]))
    reads.append(read)
```

```python
concentrations = np.array([  0.00000000e+00,   8.00000000e-09,   1.75e-08,
         3.83e-08,   8.37e-08,   1.83e-07,
         4e-07,   8.75e-07,   1.91e-06,
         4.18e-06,   9.15e-06,   2e-05])
```

```python
bos_prot = np.zeros(12)
bos_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'A%s' % index

    bos_prot[well] = reads[0]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'B%s' % index
    bos_buff[well] = reads[0]['ex280_scan_top_gain100'][buff_well_name]['480']
    
Erl_prot = np.zeros(12)
Erl_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'C%s' % index

    Erl_prot[well] = reads[1]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'D%s' % index
    Erl_buff[well] = reads[1]['ex280_scan_top_gain100'][buff_well_name]['480']
    
Gef_prot = np.zeros(12)
Gef_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'E%s' % index
    Gef_prot[well] = reads[2]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'F%s' % index
    Gef_buff[well] = reads[2]['ex280_scan_top_gain100'][buff_well_name]['480']

Pon_prot = np.zeros(12)
Pon_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'G%s' % index
    Pon_prot[well] = reads[3]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'H%s' % index
    Pon_buff[well] = reads[3]['ex280_scan_top_gain100'][buff_well_name]['480']
```

```python
wtfilenames = [
    '3rd_read_Abl_WT_Bos_20180808_210207.xml', 
    '3rd_read_Abl_WT_Erl_20180808_213647.xml', 
    '3rd_read_Abl_WT_Gef_20180808_220314.xml', 
    '3rd_read_Abl_WT_Pon_20180808_223342.xml'
]

wtdatapath = 'data/WT'
label = 'ex280_scan_top_gain100'

wtreads = list()
nreads = 4
for read in range(nreads):
    read = platereader.read_icontrol_xml(os.path.join(wtdatapath, wtfilenames[read]))
    wtreads.append(read)

wtbos_prot = np.zeros(12)
wtbos_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'A%s' % index
    wtbos_prot[well] = wtreads[0]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'B%s' % index
    wtbos_buff[well] = wtreads[0]['ex280_scan_top_gain100'][buff_well_name]['480']

wtErl_prot = np.zeros(12)
wtErl_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'C%s' % index

    wtErl_prot[well] = wtreads[1]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'D%s' % index
    wtErl_buff[well] = wtreads[1]['ex280_scan_top_gain100'][buff_well_name]['480']
    
wtGef_prot = np.zeros(12)
wtGef_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'E%s' % index
    wtGef_prot[well] = wtreads[2]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'F%s' % index
    wtGef_buff[well] = wtreads[2]['ex280_scan_top_gain100'][buff_well_name]['480']

wtPon_prot = np.zeros(12)
wtPon_buff = np.zeros(12)
ncol = 12
for well in range(ncol):
    index = well + 1
    prot_well_name = 'G%s' % index
    wtPon_prot[well] = wtreads[3]['ex280_scan_top_gain100'][prot_well_name]['480']
    buff_well_name = 'H%s' % index
    wtPon_buff[well] = wtreads[3]['ex280_scan_top_gain100'][buff_well_name]['480']
```

```python
current_palette = sns.color_palette()

wavelength = '480'
protein_color = current_palette[0]
buffer_color = current_palette[3]
lig_color = 'black'

fig = plt.figure(figsize=(10,8), dpi=300)
ax1 = fig.add_subplot(2,2,1)

# Plot fluorescence
plt.loglog(concentrations, np.flip(bos_prot, axis=0), c='black', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(bos_buff, axis=0), c='gray', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtbos_prot, axis=0), c='crimson', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtbos_buff, axis=0), c='lightpink', marker='.', linestyle='None', markersize=12)
ax1.set_ylim(100, 100000)

ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_ticks([])
plt.legend(['I403T Abl + ligand', 'ligand', 'WT Abl + ligand', 'ligand'], loc=2, prop={'size': 8})
plt.title('Bosutinib')

ax1 = fig.add_subplot(2,2,2)

# Plot fluorescence
plt.loglog(concentrations, np.flip(Erl_prot, axis=0), c='black', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(Erl_buff, axis=0), c='gray', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtErl_prot, axis=0), c='crimson', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtErl_buff, axis=0), c='lightpink', marker='.', linestyle='None', markersize=12)
ax1.set_ylim(100, 100000)

ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
plt.legend(['I403T Abl + ligand', 'ligand', 'WT Abl + ligand', 'ligand'], loc=2, prop={'size': 8})
plt.title('Erlotinib')

ax1 = fig.add_subplot(2,2,3)
# Plot fluorescence
plt.loglog(concentrations, np.flip(Gef_prot, axis=0), c='black', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(Gef_buff, axis=0), c='gray', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtGef_prot, axis=0), c='crimson', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtGef_buff, axis=0), c='lightpink', marker='.', linestyle='None', markersize=12)
ax1.set_ylim(100, 100000)

ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_tick_params(labelsize=15)
plt.legend(['I403T Abl + ligand', 'ligand', 'WT Abl + ligand', 'ligand'], loc=2, prop={'size': 8})
plt.title('Gefitinib')
    

# Plot fluorescence
ax1 = fig.add_subplot(2,2,4)
plt.loglog(concentrations, np.flip(Pon_prot, axis=0), c='black', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(Pon_buff, axis=0), c='gray', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtPon_prot, axis=0), c='crimson', marker='.', linestyle='None', markersize=12)
plt.loglog(concentrations, np.flip(wtPon_buff, axis=0), c='lightpink', marker='.', linestyle='None', markersize=12)
ax1.set_ylim(100, 100000)
ax1.yaxis.set_ticks([])

ax1.xaxis.set_tick_params(labelsize=15)
plt.legend(['I403T Abl + ligand', 'ligand', 'WT Abl + ligand', 'ligand'], loc=2, prop={'size': 8})
plt.title('Ponatinib')
    



fig.savefig('I403TAbl_%s.pdf' % wavelength, dpi=300)
```
