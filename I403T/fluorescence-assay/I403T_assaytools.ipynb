```python
import pymc
from pymc import MCMC
import assaytools
from pymc.Matplot import plot
import matplotlib 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
db = pymc.database.pickle.load('I403T Abl-Ponatinib-GH_mcmc-2018-08-10 13:37.pickle')
```

```python
ntraces = len(db.trace_names[0])
```

```python
sns.set_style('white')
sns.set_palette('muted', 12)
fig = plt.figure(figsize=(6,4*ntraces))
for i,name in enumerate(db.trace_names[0]):
    plt.subplot(ntraces,1,i+1)
    plt.plot(range(len(db.trace(name)[:])), db.trace(name)[:])
    plt.title(name)
fig.savefig('I403TAbl_ponatinib.pdf', dpi=500, bbox_inches='tight')
```
